```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
N = 17854
I0 = 199
R0 = 35
S0 = N - I0 - R0

a = 0.01
b = 0.02

t0 = 0
tmax = 200
dt = 0.01

t = np.arange(t0, tmax, dt)

def syst(t, x):
    return [0, -b * x[1], b * x[1]]

y = solve_ivp(syst, [t0, tmax], [S0, I0, R0], t_eval = t)

plt.plot(t, y.y.T)
```

```python
N = 17854
I0 = 199
R0 = 35
S0 = N - I0 - R0

a = 0.01
b = 0.02

t0 = 0
tmax = 200
dt = 0.01

t = np.arange(t0, tmax, dt)

def syst(t, x):
    return [-a * x[0], a * x[0] - b * x[1], b * x[1]]

y = solve_ivp(syst, [t0, tmax], [S0, I0, R0], t_eval = t)

plt.plot(t, y.y.T)
```
