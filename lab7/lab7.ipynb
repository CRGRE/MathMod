```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
N = 761
N0 = 6

a1 = 0.82
a2 = 0.00003

t = np.arange(0, 10, 0.1)

def d(t, x):
    return (a1 + a2 * x) * (N - x)

D = odeint(d, N0, t)

plt.plot(t, D)
```

```python
N = 761
N0 = 6

def a1(t):
    return 0.2 * np.sin(t)
def a2(t):
    return 0.8 * np.cos(t)

t = np.arange(0, 0.1, 0.01)

def d(t, x):
    return (a1(t) + a2(t) * x) * (N - x)

D = odeint(d, N0, t)

plt.plot(t, D)
```
