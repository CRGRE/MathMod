```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
M10 = 7.2
M20 = 9.1
pcr = 42
N = 45
q = 1
t1 = 28
t2 = 22
p1 = 8.1
p2 = 10.5

a1 = pcr / (t1 * t1 * p1 * p1 * N * q)
a2 = pcr / (t2 * t2 * p2 * p2 * N * q)
b = pcr / (t1 * t1 * p1 * p1 * t2 * t2 * p2 * p2 * N * q)
c1 = (pcr - p1) / (t1 * p1)
c2 = (pcr - p2) / (t2 * p2)

t0 = 0
tmax = 30
dt = 0.01
t = np.arange(t0, tmax, dt)

def syst(theta, M):
    return [M[0] - b / c1 * M[0] * M[1] - a1 / c1 * M[0] * M[0], c2 / c1 * M[1] - b / c1 * M[0] * M[1] - a2 / c1 * M[1] * M[1]]

M = solve_ivp(syst, [t0, tmax], [M10, M20], t_eval = t)

plt.plot(t, M.y.T)
```
