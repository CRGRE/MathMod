```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
a = 0.38
b = 0.043
c = 0.36
d = 0.052

x0 = [6, 23]

t0 = 0
tmax = 100
dt = 0.05

t = np.arange(t0, tmax, dt)

def syst(t, x):
    return [-a * x[0] + b * x[0] * x[1], c * x[1] - d * x[0] * x[1]]

y = solve_ivp(syst, [t0, tmax], x0, t_eval = t)

plt.plot(t, y.y.T)
```

```python
plt.plot(y.y[0], y.y[1])
plt.plot(c / d, a / b, "ro")
```
