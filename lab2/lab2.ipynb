```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
s = 20.2
n = 5.1

def f(theta, r):
    return r / np.sqrt(n ** 2 - 1)

t1 = np.arange(0, 4 * np.pi, 0.05)
x1 = s / (n + 1)
p1 = odeint(f, x1, t1)

_, ax1 = plt.subplots(subplot_kw={'projection': 'polar'})
ax1.plot(t1, p1)
```

```python
t2 = np.arange(-np.pi, 3 * np.pi, 0.05)
x2 = s / (n - 1)
p2 = odeint(f, x2, t2)

_, ax2 = plt.subplots(subplot_kw={'projection': 'polar'})
ax2.plot(t2, p2)
```
