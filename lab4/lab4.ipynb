```python
import numpy as np
import matplotlib.pyplot as plt
from scipy. integrate import odeint
```

```python
w2 = 3.7

x0 = [0.1, 1.1]

t0 = 0
tmax = 63
dt = 0.05

t = np.arange(t0, tmax, dt)

def d(x, t):
    return [x[1], -w2 * x[0]]

D = odeint(d, x0, t)

x = D[:,0]
y = D[:,1]

plt.plot(t, x)
```

```python
plt.plot(x, y)
```

```python
w2 = 10
g = 3

x0 = [0.1, 1.1]

t0 = 0
tmax = 63
dt = 0.05

t = np.arange(t0, tmax, dt)

def d(x, t):
    return [x[1], -w2 * x[0] - g * x[1]]

D = odeint(d, x0, t)

x = D[:,0]
y = D[:,1]

plt.plot(t, x)
```

```python
plt.plot(x, y)
```

```python
w2 = 11
g = 3

def f(t):
    return 0.9 * np.sin(0.9 * t)

x0 = [0.1, 1.1]

t0 = 0
tmax = 63
dt = 0.05

t = np.arange(t0, tmax, dt)

def d(x, t):
    return [x[1], -w2 * x[0] - g * x[1] - f(t)]

D = odeint(d, x0, t)

x = D[:,0]
y = D[:,1]

plt.plot(t, x)
```

```python
plt.plot(x, y)
```
